```python
import serial
import time
from datetime import datetime

# === CONFIGURAÇÃO ===
PORTA_SERIAL = 'COM6'  # Altere para 'COMx' no Windows ou '/dev/ttyUSBx' no Linux/Mac
BAUDRATE = 115200
DURACAO_SEGUNDOS = 60  # Tempo de aquisição

# Nome do arquivo de saída
nome_arquivo = f"ecg_arduino_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"

# === INÍCIO ===
print(f"Iniciando leitura serial em {PORTA_SERIAL} por {DURACAO_SEGUNDOS} segundos...")
try:
    with serial.Serial(PORTA_SERIAL, BAUDRATE, timeout=1) as ser, open(nome_arquivo, 'w') as f:
        inicio = time.time()
        while time.time() - inicio < DURACAO_SEGUNDOS:
            linha = ser.readline().decode(errors='ignore').strip()
            if linha.isdigit():
                f.write(linha + '\n')
except Exception as e:
    print(f"Erro: {e}")
else:
    print(f"Dados salvos em: {nome_arquivo}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Carrega os dados salvos
with open('ecg_arduino_20250411_230831.txt', 'r') as f:
    linhas = f.readlines()

# Filtra apenas linhas numéricas válidas (evita ruídos ou linhas vazias)
valores = [int(linha.strip()) for linha in linhas if linha.strip().isdigit()]

# Converte para numpy array
ecg = np.array(valores)

# Para Arduino com referência de 3.3V e 10 bits (0–1023)
ecg_mv = ecg / 1023.0 * 3.3  # Resultado em volts (ou mV se multiplicar por 1000)

ecg_norm = (ecg - np.mean(ecg)) / np.std(ecg)

ecg = ecg_mv

# Plota o sinal
plt.figure(figsize=(15, 4))
plt.plot(ecg[:1000])  # Exibe os primeiros 1000 pontos
plt.title('ECG capturado com Arduino + AD8232')
plt.xlabel('Amostras (~977 Hz)')
plt.ylabel('Amplitude (ADC)')
plt.grid()
plt.show()
```
